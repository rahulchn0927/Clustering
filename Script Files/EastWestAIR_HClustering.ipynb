```python
from sklearn.preprocessing import normalize
airline_norm=pd.DataFrame(normalize(airline1),columns=airline1.columns)
airline_norm
```

```python
# create dendrogram
plt.figure(figsize=(12,9))
dendrogram = sch.dendrogram(sch.linkage(airline_norm, method='complete'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=5, affinity = 'euclidean', linkage = 'ward')
```

```python
# save clusters for chart
y_hc = hc.fit_predict(airline_norm)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
#elbow method to determine no. of clusters
```

```python
Sum_of_squared_distances =[]
K = range(2,12)
for k in K:
    km =KMeans(n_clusters =k)
    km =km.fit(airline_norm)
    Sum_of_squared_distances.append(km.inertia_)
    
plt.figure(figsize=(10,7))

plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('Elbow Method For Optimal k')
plt.show()
```
